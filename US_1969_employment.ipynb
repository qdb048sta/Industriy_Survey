```python
import pandas as pd
us_df=pd.read_csv("us_employee_1970.csv",index_col=0)
sort_industry=[i for i in set(us_df["detailed_industry"])]
sort_industry.sort()
sort_gender=[i for i in set(us_df["gender"])]
sort_gender.sort()
empty_gender=[]
male=[]
female=[]
for i in sort_industry:
    for g in sort_gender:
        counted=us_df[(us_df["detailed_industry"]==i) & (us_df["gender"]==g)].count()[0]
        if g==" ":
            empty_gender.append(counted)
        elif g=="1":
            male.append(counted)
        elif g=="2":
            female.append(counted)
counted_us=pd.DataFrame()
counted_us["empty"]=empty_gender
counted_us["male"]=male
counted_us["feamle"]=female
counted_us.index=sort_industry


        
```
