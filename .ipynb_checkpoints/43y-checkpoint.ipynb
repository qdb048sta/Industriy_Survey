```python
df=pd.read_csv("D:\\User_Data\\Downloads\\food_production.csv")#food_production is acutally the total manufacturing info
```

```python
subdfs=df[["Unnamed: 2","Unnamed: 3"]][8:30]
subdfs.index=[i for i in range(0,22)]
actual_col=[]
for i in range(19,27+1):
    subdf=df[["Unnamed: 2","Unnamed: 3"]][8+24*i:30+24*i]
    subdf.index=[i for i in range(0,22)]
    subdf.columns=[str(i)+"M",str(i)+"F"]
    actual_col.append(str(i)+"M")
    actual_col.append(str(i)+"F")
    print(subdfs)
    subdfs=pd.concat([subdfs,subdf],axis=1)
    subdfs=subdfs.replace(to_replace='-',value=0)
M=[]
F=[]
for i in range(0,22):
    for g in ["M","F"]:
        summ=0
        for c in range(19,28):
            v=subdfs[str(c)+g][i]
            if v=="-":
                value=0
            else:
                value=int(v)
            summ=value+summ
        if g=="M":
            M.append(summ)
        elif g=="F":
            F.append(summ)

        print(str(i)+"_"+str(g))
        print(summ)
new_df=pd.DataFrame()
new_df["M"]=M
new_df["F"]=F
male=[m for m in subdfs.columns if "M" in m]
malesubdfs=pd.DataFrame()
for m in male:
    malesubdfs[m]=pd.to_numeric(subdfs[m])
sum_m=malesubdfs.T.sum()
female=[f for f in subdfs.columns if "F" in f]
femalesubdfs=pd.DataFrame()
for f in female:
    femalesubdfs[f]=pd.to_numeric(subdfs[f])
sum_f=femalesubdfs.T.sum()
sum_df=pd.DataFrame([sum_m,sum_f])
sum_df.T.to_csv("43textile.csv")
```
